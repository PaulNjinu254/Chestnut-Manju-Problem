```python
import matplotlib.pyplot as plt
import math
import time
start = time.time()

def time_to_fill_space(initial_volume, target_volume, doubling_time=1):
    volume = initial_volume
    time = 0
    times = [0]
    volumes = [volume]

    while volume < target_volume:
        time += doubling_time
        volume *= 2
        times.append(time)
        volumes.append(volume)

    return time, times, volumes

# Constants
R_bun = 0.05  # radius in meters
V_bun = (4/3) * math.pi * R_bun**3  # ~0.000524 m^3

R_solar = 100 * 1.496e11  # 100 AU in meters
V_solar = (4/3) * math.pi * R_solar**3  # ~1.4e40 m^3

# Run simulation
minutes_needed, time_steps, volume_steps = time_to_fill_space(V_bun, V_solar)

#Plotting the growth
plt.figure(figsize=(10,6))
plt.plot(time_steps, volume_steps, label="Chestnut Bun Volume", color='brown')
plt.axhline(V_solar, color='red', linestyle='--', label="Solar System Volume")
plt.yscale('log')
plt.xlabel("Time (minutes)")
plt.ylabel("Total Volume (m)")
plt.title("Chestnut Buns Filling the Solar System Over Time")
plt.legend()
plt.show()

#Output
print(f"\n It will take approximately {minutes_needed} minutes for chestnut buns to fill the solar system.")

elapsed_time = time.time() - start
print("time to run code is: {} seconds".format(elapsed_time))
```
